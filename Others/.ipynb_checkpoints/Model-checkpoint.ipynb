## Bike rental demand prediction using Machine Learning

### Building Machine Learning Models
Now we will train several Machine Learning models and compare their results.

```python
# Import necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
from sklearn import metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
import warnings
warnings.simplefilter(action = 'ignore', category = FutureWarning)
plt.rcParams['figure.figsize'] = (12.0, 5.0)
```

```python
# Read data
df = pd.read_csv('../data/train.csv', header = 0, error_bad_lines=False)
```

```python
# Import the preprocessing from utils
from ipynb.fs.full.utils import preprocessing
df = preprocessing(df)
```

```python
df.head()
```

### Train_Test_Split

```python
# Separate Target Variable and Predictor Variables
X = df.drop("count", axis=1)
y = df["count"]

# Split dataset so that we can train the model with the training set and test the predictive power with the test dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)

# Normalize the predictors (X_values) using MinMaxScaler
mm = MinMaxScaler()
X_train = mm.fit_transform(X_train)
X_test = mm.transform(X_test)
```

```python
X_test.shape, X_train.shape, y_test.shape, y_train.shape
```

### Metrics for evaluating the models

```python
# Metrics for evaluating the models
# code source: https://github.com/Akshaypatil7/

def rmsle(y,y_):
    log1 = np.nan_to_num(np.array([np.log(v + 1) for v in y]))
    log2 = np.nan_to_num(np.array([np.log(v + 1) for v in y_]))
    calc = (log1 - log2) ** 2
    return np.sqrt(np.mean(calc))
def scores(y, y_):
    print('r square  ', metrics.r2_score(y, y_))
    print('Adjusted r square:{}'.format(1 - (1-metrics.r2_score(y, y_))*(len(y)-1)/(len(y)-X_train.shape[1]-1)))
    print('MAPE:{}'.format(np.mean(np.abs((y - y_) / y))*100))
    print('MSE:', metrics.mean_squared_error(y, y_))
    print('RMSE:', np.sqrt(metrics.mean_squared_error(y, y_))) 
def test_scores(model):
    print('<<<------------------- Training Data Score --------------------->')
    print()
    #Predicting result on Training data
    y_pred = model.predict(X_train)
    scores(y_train,y_pred)
    print('RMSLE:',rmsle(y_train,y_pred))
    print()
    print('<<<------------------- Test Data Score --------------------->')
    print()
    # Evaluating on Test Set
    y_pred = model.predict(X_test)
    scores(y_test,y_pred)
    print('RMSLE:',rmsle(y_test,y_pred))
```

```python
def model_performance(model):
    model.fit(X_train, y_train)
    print('\033[1m' + '\033[94m' + str(model) +' Evaluation Scores'.center(80) + '\033[0m')
    test_scores(model)
    return test_scores
    
```

### Linear Regression

```python
# Linear Regression
LR = LinearRegression()
model_performance(LR)
```

### Decision Tree Regressor

```python
#Decision Tree Regressor
DT = DecisionTreeRegressor()
model_performance(DT)
```

### Random Forest Regressor

```python
#Random Forest Regressor before Hyperparameter Tuning
RF = RandomForestRegressor()
model_performance(RF)
```

### K Nearest Neighbor

```python
#K Nearest Neighbor before Hyperparameter Tuning
KNN = KNeighborsRegressor() 
model_performance(KNN)
```

### XGBoost Regressor

```python
#XGB Regressor before Hyperparameter Tuning
XGB = XGBRegressor()
model_performance(XGB)
```

### Hyperparameter Optimization using GridSearchCV

```python
# function for Hyperparameter Optimization using GridSearchCV
def Grid(model, param_grid):

    # Instantiate the gridSearchCV object: reg_cv
    grid_cv = GridSearchCV(model, param_grid, cv=5,scoring='r2')

    # Fit it to the data
    grid_cv.fit(X, y)

    # Print the tuned parameters and score
    print("Tuned Decision reg Parameters: {}".format(grid_cv.best_params_))
    print("Best score is {}".format(grid_cv.best_score_))
    return model
```

### Random Forest Regressor

```python
RF.get_params()
```

```python
#Random Forest Regressor with Tuned Hyperparameters
RFR = RandomForestRegressor(max_depth= 15)
RFR.fit(X_train, y_train)
print('\033[1m' + '\033[94m' +'Score After Hyperparameter Tuning '.center(80) + '\033[0m')
test_scores(RFR)
```

### XGB Regressor

```python
#XGB Regressor with Tuned Hyperparameters
XGBr = XGBRegressor(max_depth=10)
XGBr.fit(X_train, y_train)
print('\033[1m' + '\033[94m' +'Score After Hyperparameter Tuning '.center(80) + '\033[0m')
test_scores(XGBr)
```

```python
def pred(model):
    ytest_pred = model.predict(X_test)
    ytest_prediction = pd.DataFrame(ytest_pred,index = y_test.index,columns=['count'])
    df_model = pd.merge(y_test, ytest_prediction, left_index=True, right_index=True, how="inner")
    print('Predicted Bike Rental Count ' +str(model))
    return df_model
```
